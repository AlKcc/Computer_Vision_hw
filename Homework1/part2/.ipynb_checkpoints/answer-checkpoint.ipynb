```python
# 本次作业你可能需要用到的包和函数

%matplotlib inline
# 导入所需模块
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2

from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# plt显示灰度图片
def plt_show(img):
    plt.imshow(img,cmap='gray')
    plt.show()

# 读取一个文件夹下的所有图片，输入参数是文件名，返回文件地址列表
def read_directory(directory_name):
    faces_addr = []
    for filename in os.listdir(directory_name):
        faces_addr.append(directory_name + "/" + filename)
    return faces_addr
```

 1. 将数据集划分为80%的训练集，20%的测试集，在训练集上使用PCA将特征维度降为100，即得到100个特征和其对应的特征向量，并使用训练得到的PCA将测试集维度也压缩到100，**输出：压缩后的训练集维度和测试集维度、经过PCA得到的特征向量维度**。

```python
# Your Code
# 读取图片数据并转换为numpy数组
def read_images(addr_list):
    images = []
    for addr in addr_list:
        img = cv2.imread(addr, 0)  # 以灰度模式读取
        img = img.flatten()  # 展平图像
        images.append(img)
    return np.array(images)

# 读取data文件夹下所有图片地址
data_directory = "data"
all_faces_addr = read_directory(data_directory)

# 读取图片数据
all_images = read_images(all_faces_addr)

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(all_images, np.arange(40).repeat(10), test_size=0.2, random_state=42)

# 在训练集上使用PCA降维
pca = PCA(n_components=100)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

print("压缩后的训练集维度:", X_train_pca.shape)
print("压缩后的测试集维度:", X_test_pca.shape)
print("经过PCA得到的特征向量维度:", pca.components_.shape)
```

2. 得到100个特征向量后，**使用这些向量重建特征脸并输出**。

```python
# 重建特征脸
# 每个特征脸是PCA的components_中的一行，需要将其还原为图像形状
img_height = 112
img_width = 92
num_components = 100
eigenfaces = pca.components_[:num_components]

# 将特征向量还原为图像形状
reconstructed_eigenfaces = []
for eigenface in eigenfaces:
    eigenface_image = eigenface.reshape((img_height, img_width))
    reconstructed_eigenfaces.append(eigenface_image)

# 展示重建的特征脸
fig, axes = plt.subplots(nrows=6, ncols=10, figsize=(12, 8))
for i, ax in enumerate(axes.flat):
    ax.imshow(reconstructed_eigenfaces[i], cmap='gray')
    ax.axis('off')
plt.show()
```

3. 使用这100个特征和对应的特征向量，自行在训练集和测试集中**分别**选择**5张**人脸图片进行人脸重建并**输出对比图**。

```python
# 训练集和测试集分别选择5张图片的索引
train_indices = np.random.choice(len(X_train_pca), 5, replace=False)
test_indices = np.random.choice(len(X_test_pca), 5, replace=False)

# 重建训练集选中的人脸
reconstructed_train_faces = []
for index in train_indices:
    # 将降维后的向量通过PCA反变换回原始维度
    reconstructed_face = pca.inverse_transform(X_train_pca[index].reshape(1, -1))
    reconstructed_face = reconstructed_face.reshape((img_height, img_width))
    reconstructed_train_faces.append(reconstructed_face)

# 重建测试集选中的人脸
reconstructed_test_faces = []
for index in test_indices:
    reconstructed_face = pca.inverse_transform(X_test_pca[index].reshape(1, -1))
    reconstructed_face = reconstructed_face.reshape((img_height, img_width))
    reconstructed_test_faces.append(reconstructed_face)

# 原始训练集选中的人脸（从原始数据中获取）
original_train_faces = X_train[train_indices].reshape(-1, img_height, img_width)
# 原始测试集选中的人脸（从原始数据中获取）
original_test_faces = X_test[test_indices].reshape(-1, img_height, img_width)

# 绘制对比图
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(12, 6))
for i in range(5):
    # 绘制训练集原始图像和重建图像
    axes[0, i].imshow(original_train_faces[i], cmap='gray')
    axes[0, i].set_title('Original Train')
    axes[0, i].axis('off')
    axes[1, i].imshow(reconstructed_train_faces[i], cmap='gray')
    axes[1, i].set_title('Reconstructed Train')
    axes[1, i].axis('off')

fig.suptitle('Training Set Face Reconstruction Comparison')
plt.show()

fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(12, 6))
for i in range(5):
    # 绘制测试集原始图像和重建图像
    axes[0, i].imshow(original_test_faces[i], cmap='gray')
    axes[0, i].set_title('Original Test')
    axes[0, i].axis('off')
    axes[1, i].imshow(reconstructed_test_faces[i], cmap='gray')
    axes[1, i].set_title('Reconstructed Test')
    axes[1, i].axis('off')

fig.suptitle('Testing Set Face Reconstruction Comparison')
plt.show()
```

4. **输出**降维后**每个**新特征向量所占的信息量占原始数据总信息量的百分比，以及所有返回特征所携带的**信息量总和**是原始数据的多少；

```python
from tabulate import tabulate

# 计算每个主成分所解释的方差比率，这就代表每个新特征向量所占的信息量占原始数据总信息量的百分比
explained_variance_ratio = pca.explained_variance_ratio_
# 计算所有返回特征所携带的信息量总和（累计解释方差比率）
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# 准备表格数据
table_data = []
for i, ratio in enumerate(explained_variance_ratio):
    percentage = f"{ratio * 100:.2f}%"
    cumulative_percentage = f"{cumulative_explained_variance[i] * 100:.2f}%"
    table_data.append([i + 1, percentage, cumulative_percentage])

# 定义表格表头 
headers = ["特征向量序号", "单个特征信息量占比", "累计信息量占比"]

# 打印表格
print(tabulate(table_data, headers=headers, tablefmt="grid"))

# 单独打印所有返回特征所携带的信息量总和是原始数据的多少
total_information_percentage = cumulative_explained_variance[-1] * 100
print(f"\n所有返回特征所携带的信息量总和是原始数据的 {total_information_percentage:.2f}%")
```

5. **画出**特征个数和所携带信息数的曲线图，此处特征数**上限设置为150**

```python
# 计算每个主成分所解释的方差比率
explained_variance_ratio = pca.explained_variance_ratio_

# 计算累计解释方差比率
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# 特征个数范围，上限设为150
n_components = np.arange(1, 151)
if len(cumulative_explained_variance) < 150:
    # 如果实际主成分不足150个，对累计解释方差比率进行补全
    num_missing = 150 - len(cumulative_explained_variance)
    cumulative_explained_variance = np.pad(cumulative_explained_variance, (0, num_missing), 'edge')

# 绘制曲线图
plt.plot(n_components, cumulative_explained_variance[:len(n_components)], 'b-')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance (Information)')
plt.title('Number of Features vs. Information Carried')
plt.show()
```

6.  保留特征数为150，得到的训练集每次保留前（n=n+1，n初值为1）个特征，分别训练一个KNN分类器，KNN分类器的参数*n_neighbors*为3，观察并验证测试集在使用其前n个特征时KNN分类的准确率。

```python
from sklearn.metrics import accuracy_score
# KNN分类器参数
n_neighbors = 3

# 存储特征保留数和对应的准确率
num_features_list = []
accuracies = []

# 进行循环，每次保留前n个特征进行训练和测试
for n in range(1, 151):
    # 取训练集和测试集的前n个特征
    X_train_subset = X_train_pca[:, :n]
    X_test_subset = X_test_pca[:, :n]

    # 初始化并训练KNN分类器
    knn = KNeighborsClassifier(n_neighbors=n_neighbors)
    knn.fit(X_train_subset, y_train)

    # 预测并计算准确率
    y_pred = knn.predict(X_test_subset)
    accuracy = accuracy_score(y_test, y_pred)

    num_features_list.append(n)
    accuracies.append(accuracy)

# 绘制不同特征保留数和准确率的曲线图
plt.plot(num_features_list, accuracies, 'b-')
plt.xlabel('Number of Features Retained')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. Number of Features Retained')
plt.show()
```
